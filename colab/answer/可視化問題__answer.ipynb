## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()

a_data_hist = pd.read_csv("hist_data.csv",usecols=[0],squeeze=True)
b_data_hist = pd.read_csv("hist_data.csv",usecols=[1],squeeze=True)
plt.hist(a_data_hist, bins=30, alpha=0.5, color='m',label="aデータ")
plt.hist(b_data_hist, bins=30, alpha=0.5, color='g',label="bデータ")
plt.legend()
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import japanize_matplotlib
japanize_matplotlib.japanize()

class_wid_len = pd.read_csv("scatter_data.csv")
class0 = class_wid_len[class_wid_len["class"] == 0]
class1 = class_wid_len[class_wid_len["class"] == 1]
class2 = class_wid_len[class_wid_len["class"] == 2]
plt.scatter(class0["length (cm)"], class0["width (cm)"],label="class0",c="r")
plt.scatter(class1["length (cm)"], class1["width (cm)"],label="class1",c="m")
plt.scatter(class2["length (cm)"], class2["width (cm)"],label="class2",c="g")
plt.title("クラス０～２のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.grid(color = 'k')
plt.legend()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(2).png

```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()

time = pd.read_csv("line_data.csv",squeeze=True,usecols=[0])
amount = pd.read_csv("line_data.csv",squeeze=True,usecols=[1])
cost = pd.read_csv("line_data.csv",squeeze=True,usecols=[2])
plt.plot(time,amount,marker="o")
plt.xticks(rotation=30)
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
plt.grid(color = 'k')
plt.show()
plt.bar(time,cost,alpha=0.7)
plt.xticks(rotation=30)
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用（円）")
plt.grid(color = 'k')
plt.show()
```
