## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()

a_data_hist = pd.read_csv("hist_data.csv",usecols=[0],squeeze=True)
b_data_hist = pd.read_csv("hist_data.csv",usecols=[1],squeeze=True)
plt.hist(a_data_hist, bins=30, alpha=0.5, color='m')
plt.hist(b_data_hist, bins=30, alpha=0.5, color='g')
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.show()
```
